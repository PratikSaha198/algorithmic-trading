```python
# Quandl

import quandl

with open("quandl_api_key.txt") as file:
    api_key = file.read()[:-1]
    
quandl.ApiConfig.api_key = api_key
```

```python
crude_oil_price = quandl.get('EIA/PET_RWTC_D')
```

```python
import matplotlib.pyplot as plt

crude_oil_price.plot(figsize=(12, 6))
```
