```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()

ax_1 = fig.add_axes([0, 0, 1, 1])
ax_2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax_1.plot(x, y, 'r')
ax_2.plot(x, y, 'r')
```

```python
fig = plt.figure()

ax_1 = fig.add_axes([0, 0, 1, 1])
ax_2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax_1.plot(x, z, 'r')
ax_1.set_xlabel('X')
ax_1.set_ylabel('Z')

ax_2.plot(x, y, 'r')
ax_2.set_title('Zoom')
ax_2.set_xlabel('X')
ax_2.set_ylabel('Y')
ax_2.set_xlim([20, 22])
ax_2.set_ylim([30, 50])
```

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y, color='b', lw=2, ls='--')
axes[1].plot(x, z, color='r', lw=2, ls='-')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(x, y, color='b', lw=2, ls='-')
axes[1].plot(x, z, color='r', lw=2, ls='--')

plt.tight_layout()
```
