```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#FF8C00')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', linewidth=3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', linewidth=3, alpha=0.1)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', linewidth=3, linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', linewidth=3, linestyle=':')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', lw=1, ls='-', marker='o', markersize=5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', 
        lw=1, ls='-', 
        marker='o', markersize=5,
        markerfacecolor='yellow', 
        markeredgewidth=3,
        markeredgecolor='green')
```

```python
fig = plt.figure()

x = np.linspace(0, 5, 100)
y = x ** 2

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='black', linewidth=3, linestyle='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
```
