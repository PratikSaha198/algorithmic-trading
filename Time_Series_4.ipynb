```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../05-Pandas-with-Time-Series/time_data/walmart_stock.csv', index_col='Date', parse_dates=True)
```

```python
df['Open'].plot()
df.rolling(7).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Open'].plot()
df.rolling(30).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Open'].plot()
df.rolling(90).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(30).mean()
df[['Close', 'Close 30 Day MA']].plot(figsize=(16, 6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16, 6))
```

```python
# Boolinger Bands

df['Close: 20 Day Mean'] = df['Close'].rolling(window=20).mean()
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())


df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16, 6))
```
