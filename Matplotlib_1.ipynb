```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Functional Method

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r-')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')

plt.show()
```

```python
# Object Oriented Method

fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title('Title')
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes_1.plot(x, y)
axes_1.set_title('Larger Plot')

axes_2.plot(y, x)
axes_2.set_title('Smaller Plot')
```
